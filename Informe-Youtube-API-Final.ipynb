## Justificación ##
El uso del API de YouTube tiene varias ventajas sobre el web scraping:

<font color="red">*Acceso estructurado a los datos:*</font> El API de YouTube proporciona una estructura bien definida para acceder a los datos de YouTube. Puedes realizar consultas específicas y recibir los resultados en un formato estructurado, como JSON, lo que facilita su manipulación y análisis.

<font color="red">*Cumplimiento de los términos de servicio:*</font> Al utilizar el API de YouTube, estás siguiendo los términos de servicio establecidos por YouTube y Google. Esto garantiza que estás utilizando los datos de YouTube de acuerdo con las reglas y restricciones establecidas por la plataforma.

<font color="red">*Actualizaciones y mantenimiento:*</font> El API de YouTube es administrado y actualizado por el equipo de desarrollo de YouTube. Esto significa que si hay cambios en la plataforma, es más probable que el API se actualice para reflejar esos cambios, lo que te permite mantener tu aplicación sincronizada con las últimas funcionalidades de YouTube.

<font color="red">*Rendimiento y eficiencia:*</font> El API de YouTube está diseñado para manejar grandes volúmenes de solicitudes y ofrecer respuestas rápidas y eficientes. Esto garantiza un mejor rendimiento y evita la sobrecarga de los servidores de YouTube.

```python
#https://www.youtube.com/watch?v=yKtBmciU3AM

import googleapiclient.discovery
import googleapiclient.errors
import urllib.parse
import matplotlib.pyplot as plt

enlace_video_principal = input("Ingresar el enlace del video: ")

# Clave de API de YouTube (reemplaza con tu propia clave)
api_key = "AIzaSyBtlNY36JrBI1ItamxPI0jLNE9CCNKkysI"

def obtener_videos_recomendados_por_enlace(enlace_video):
    try:
        # Obtener el ID del video principal a partir del enlace
        video_id = urllib.parse.urlparse(enlace_video).query.split('=')[1]

        # Crear una instancia del cliente de la API de YouTube
        youtube = googleapiclient.discovery.build('youtube', 'v3', developerKey=api_key)

        # Llamar a la API de YouTube para obtener los videos recomendados
        response = youtube.search().list(part='snippet', relatedToVideoId=video_id, type='video', maxResults=10).execute()
        videos_recomendados = response['items']

        # Obtener los títulos y números de likes de los videos recomendados
        titulos = []
        likes = []
        
        # Abrir el archivo para escribir los resultados
        with open("videos_recomendados.txt", "w", encoding="utf-8") as file:
            # Escribir los detalles de los videos recomendados en el archivo
            file.write("Videos recomendados:\n")
            for video in videos_recomendados:
                # Obtener el ID del video recomendado
                video_id = video['id']['videoId']
                
         

                # Llamar a la API de YouTube para obtener estadísticas del video recomendado
                video_response = youtube.videos().list(part='snippet,statistics', id=video_id).execute()
                video_info = video_response['items'][0]

                # Obtener los detalles del video recomendado
                titulo = video_info['snippet']['title']
                descripcion = video_info['snippet']['description']
                like_count = int(video_info['statistics']['likeCount'])
                comentarios = video_info['statistics']['commentCount']
                enlace = f"https://www.youtube.com/watch?v={video_id}"
                
                # Escribir los detalles del video recomendado en el archivo
                file.write("Título: {}\n".format(titulo))
                file.write("Descripción: {}\n".format(descripcion))
                file.write("Número de likes: {}\n".format(like_count))
                file.write("Número de comentarios: {}\n".format(comentarios))
                file.write("Enlace: {}\n".format(enlace))
                file.write("\n")
                
                 # Imprimir los detalles del video recomendado
                print("Título:", titulo)
                #print("Descripción:", descripcion)
                print("Número de likes:", like_count)
                print("Número de comentarios:", comentarios)
                print("Enlace:", enlace)
                print()

                # Agregar el título y número de likes a las listas correspondientes
                if like_count > 7000:  # Más de 10,000 likes
                    titulos.append(titulo)
                    likes.append(like_count)


        # Personalizar colores para las barras
        yt = ['#282828','#606060','#E0E0E0','#FF0000','#CD0000']
        # Crear el gráfico de barras con colores personalizados
        plt.bar(titulos, likes, color=yt)
        plt.xlabel('Títulos de videos')
        plt.ylabel('Número de likes')
        plt.title('Videos recomendados por título y número de likes')
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()  
        
    except googleapiclient.errors.HttpError as e:
        print("Error HTTP al llamar a la API de YouTube:", e)
    except googleapiclient.errors.Error as e:
        print("Error al llamar a la API de YouTube:", e)

# Ejemplo de uso: obtener videos recomendados a partir del enlace del video principal
obtener_videos_recomendados_por_enlace(enlace_video_principal)
```

### <span style="color:#EE0000">1. Comentarios con mayor número de likes y nube de palabras </span>

```python
#https://youtu.be/yKtBmciU3AM

import googleapiclient.discovery
import googleapiclient.errors
import urllib.parse
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

enlace_video = input("Ingresar enlace del video: ")

# Clave de API de YouTube (reemplaza con tu propia clave)
api_key = "AIzaSyBtlNY36JrBI1ItamxPI0jLNE9CCNKkysI"

def obtener_comentarios_mas_populares(enlace_video, cantidad):
    try:
        # Obtener el ID del video a partir del enlace
        video_id = urllib.parse.urlparse(enlace_video).path.split('/')[1]

        # Crear una instancia del cliente de la API de YouTube
        youtube = googleapiclient.discovery.build('youtube', 'v3', developerKey=api_key)

        # Llamar a la API de YouTube para obtener los comentarios del video
        response = youtube.commentThreads().list(part='snippet', videoId=video_id, order='relevance', textFormat='plainText').execute()
        comentarios = response['items']

        # Ordenar los comentarios por número de likes de mayor a menor
        comentarios_ordenados = sorted(comentarios, key=lambda x: int(x['snippet']['topLevelComment']['snippet']['likeCount']), reverse=True)

        # Obtener los comentarios más populares
        comentarios_populares = comentarios_ordenados[:cantidad]

        # Guardar los comentarios en un archivo
        with open("comentarios.txt", "w", encoding="utf-8") as file:
            file.write(f"Los {cantidad} comentarios con más likes:\n")
            for i, comentario in enumerate(comentarios_populares, start=1):
                likes = comentario['snippet']['topLevelComment']['snippet']['likeCount']
                texto_comentario = comentario['snippet']['topLevelComment']['snippet']['textDisplay']
                file.write(f"Likes: {likes}\n")
                file.write(f"Comentario: {texto_comentario}\n")
                file.write("\n")

        print("Los comentarios se han guardado en el archivo 'comentarios.txt'.")
        
        # Imprimir los comentarios más populares
        print(f"Los {cantidad} comentarios con más likes:")
        for i, comentario in enumerate(comentarios_populares, start=1):
            likes = comentario['snippet']['topLevelComment']['snippet']['likeCount']
            texto_comentario = comentario['snippet']['topLevelComment']['snippet']['textDisplay']
            #print(f"Comentario {i}:")
            print(f"Likes: {likes}")
            print(f"Comentario: {texto_comentario}")
            
            print()


        # Obtener las palabras vacías en español
        stopwords_es = set(stopwords.words('spanish'))

        # Generar el texto para la nube de palabras
        texto_comentarios = ''
        for comentario in comentarios_populares:
            texto = comentario['snippet']['topLevelComment']['snippet']['textDisplay']
            texto_comentarios += texto + ' '

        # Generar la nube de palabras
        wordcloud = WordCloud(stopwords=stopwords_es, width=800, height=400, background_color='white', colormap='autumn').generate(texto_comentarios)

        # Mostrar la nube de palabras
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.show()

    except googleapiclient.errors.HttpError as e:
        print("Error HTTP al llamar a la API de YouTube:", e)
    except googleapiclient.errors.Error as e:
        print("Error al llamar a la API de YouTube:", e)

# Ejemplo de uso: obtener los 5 comentarios con más likes a partir del enlace de un video

cantidad_comentarios = 5  # Cantidad de comentarios a obtener
obtener_comentarios_mas_populares(enlace_video, cantidad_comentarios)
```

```python
#Credenciales del API de youtube
key = 'AIzaSyBSAPjHK5OT2Xta0RELxmPJHnOfZQjMq_A'

# Crear una instancia del cliente de la API de YouTube
youtube = build('youtube','v3',developerKey = key)

def videos_csv(youtube,canal):
    canal = canal.split(",")
    canales = []
    solicitud = youtube.channels().list(
    part = 'snippet,contentDetails,statistics',
    id = canal)
    respuesta = solicitud.execute()
    for i in range(len(respuesta['items'])):
        datos = dict(nombre_canal = respuesta['items'][i]['snippet']['title'],
                 suscriptores = int(respuesta['items'][i]['statistics']['subscriberCount']),
                 vistas = int(respuesta['items'][i]['statistics']['viewCount']),
                 numero_videos = int(respuesta['items'][i]['statistics']['videoCount']),
                 id_playlist = respuesta['items'][i]['contentDetails']['relatedPlaylists']['uploads'])
        canales.append(datos)
    contador = 0    
    canales = pd.DataFrame(canales)
    for ID in canales['id_playlist']:
        nombre_canal = canales['nombre_canal'].iloc[contador] + '.csv'
        solicitud = youtube.playlistItems().list(
                part='contentDetails',
                playlistId=ID,
                maxResults=50)
        respuesta = solicitud.execute()
        id_videos =[]
        for i in range(len(respuesta['items'])):
            id_videos.append(respuesta['items'][i]['contentDetails']['videoId'])
        
        pasar_paginas = respuesta['nextPageToken']
        mas_paginas = True
        
        while mas_paginas:
            if pasar_paginas is None:
                mas_paginas = False
            else: 
                solicitu_d = youtube.playlistItems().list(
                            part = 'contentDetails',
                            playlistId = ID,
                            maxResults = 50,
                            pageToken = pasar_paginas)
                respuest_a = solicitu_d.execute()
                for i in range(len(respuest_a['items'])):
                    id_videos.append(respuest_a['items'][i]['contentDetails']['videoId'])
                
                pasar_paginas = respuest_a.get('nextPageToken')
        
        estadisticas_videos = []
        for i in range(0,len(id_videos),50):
            solicitud = youtube.videos().list(
                        part = 'snippet, statistics',
                        id = ','.join(id_videos[i:i+50]))
            respuesta = solicitud.execute()
            for j in respuesta['items']:
                estadisticas = dict(Titulo = j['snippet']['title'],
                               fecha_publicacion = j['snippet']['publishedAt'],
                               Vistas = int(j['statistics'].get('viewCount',0)),
                               Likes = int(j['statistics'].get('likeCount',0)),
                               Dislikes = j['statistics'].get('dislikeCount',0),
                               Comentarios = int(j['statistics'].get('commentCount',0)))
                estadisticas_videos.append(estadisticas)
        estadisticas_videos = pd.DataFrame(estadisticas_videos)
        with open(nombre_canal,'w') as archivo:
            estadisticas_videos.to_csv(archivo, index = False)
        contador +=1
        print(contador)
    return 'Creado satisfactoriamente'
```
